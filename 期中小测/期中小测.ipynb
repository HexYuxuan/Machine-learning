```python
import numpy as np
import pandas
import random
import string

def find_friend(person):
    # 创建空字典
    friends_dict = {}

    # 遍历每个人
    for i in range(len(person)):
        # 将数字值转换为浮点数
        person[i][1] = float(person[i][1])

        # 初始化朋友列表
        friends_list = []

        # 排序数组，按数字值的绝对差进行排序
        sorted_person = sorted(person, key=lambda x: abs(float(x[1]) - float(person[i][1])))

        # 获取与当前人数字值最接近的5个人
        for j in range(1, 6):
            friends_list.append(sorted_person[j][0])

        # 添加到朋友字典中
        friends_dict[person[i][0]] = friends_list

    return friends_dict
    
# for example 生成随机数组用于测试
person_test1 = np.array([chr(i) for i in range(65, 91)])
person_test2 = np.zeros(26)
for i in range(0,26):
    person_test2[i] = i + random.randrange(-26,26)

person_test = np.column_stack((person_test1, person_test2))

friends = find_friend(person_test)

print(person_test)
for key, value in friends.items():
    print(key, value)
```

### 1.原始数据导入

```python
import pandas as pd

# 读取CSV文件
file_path = 'data_new.csv'
data = pd.read_csv(file_path)

# 显示数据
data.head()
```

### 2.初始数据探索

(1) 计算每个定量属性的平均值、标准偏差、最小值和最大值。

```python
from pandas.api.types import is_numeric_dtype

for col in data.columns:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Mean = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())
```

### 2.数据转换

我们可以看到，对于地区，目标结果，他们的数据类型都是字符串，不便于进行学习。所以首先我们需要把数据中的字符串替换成数字，然后在需要得倒结果时再根据对应表得到结果。
首先，我们来将地区转化为其对应的数字，由于地区的类型很多，所以我们先打印出所以地区类型及其对应的数字

```python
# 得倒地区数据
district = data.iloc[:, 0]
# 初始化一个字典来存储不同字符串和它们对应的数字
unique_dict = {}
current_number = 1

# 遍历地区数据
for value in district:
    if value not in unique_dict:
        unique_dict[value] = current_number
        print(f'{value}: {current_number}')
        current_number += 1
```

接着我们来替换字符串，

```python
# 使用字典替换地区数据中的字符串为对应的数字
data['district'] = data['district'].map(unique_dict)

data.head()
```

接下来我们来获取目标类型

```python
# 打印最后一列数据
target = data.iloc[:, -1]

# 初始化一个字典来存储最后一列不同字符串和它们对应的数字
unique_target = {}
current_number = 1

# 遍历最后一列数据
for value in target:
    if value not in unique_target:
        unique_target[value] = current_number
        print(f'{value}: {current_number}')
        current_number += 1
```

接着我们来替代为数字

```python
# 使用字典替换最后一列数据中的字符串为对应的数字
data['target'] = data['target'].map(unique_target)

# 显示修改后的数据前几行
data.head()
```

（3） 显示每个属性的直方图

```python
import matplotlib.pyplot as plt
# 确定子图的布局
num_columns = len(data.columns)
num_rows = (num_columns + 1) // 2  # 两行布局

fig, axes = plt.subplots(num_rows, 2, figsize=(15, num_rows * 5))

# 遍历每一列并绘制直方图
for i, column in enumerate(data.columns):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    ax.hist(data[column], bins=30, edgecolor='black')
    ax.set_title(f'Histogram of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')

# 如果图的数量是奇数，删除多余的子图
if num_columns % 2 != 0:
    fig.delaxes(axes[-1, -1])

plt.tight_layout()
plt.show()
```

### 4.数据预处理
（1） 查看数据丢失

```python
import numpy as np

data = data.replace('?',np.NaN)

print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))

print('Number of missing values:')
for col in data.columns:
    print('\t%s: %d' % (col,data[col].isna().sum()))
```

可以看到这个数据集数据比较饱满，没有对视的数据

（2） 寻找重复数据，并修正
首先来查看是否存在重复样本

```python
dups = data.duplicated()
print('Number of duplicate rows = %d' % (dups.sum()))
```

这里我们可以看到有48个重复样本，然后我们将这些重复样本设置为一个样本，这里我们可以查看修正前后的样本数量

```python
print('Number of rows before discarding duplicates = %d' % (data.shape[0]))
data2 = data.drop_duplicates()
print('Number of rows after discarding duplicates = %d' % (data2.shape[0]))
```
